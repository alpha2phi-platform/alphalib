# EDA

```python
DATASET_FILE = "alphalib/data/MDTaug13.xlsx"
COLUMNS = ["year",
           "month",
           "day",
           "first_prize",
           "second_prize",
           "third_prize",
           "starter_prize_1",
           "starter_prize_2",
           "starter_prize_3",
           "starter_prize_4",
           "starter_prize_5",
           "starter_prize_6",
           "starter_prize_7",
           "starter_prize_8",
           "starter_prize_9",
           "starter_prize_10",
           "consolation_prize_1",
           "consolation_prize_2",
           "consolation_prize_3",
           "consolation_prize_4",
           "consolation_prize_5",
           "consolation_prize_6",
           "consolation_prize_7",
           "consolation_prize_8",
           "consolation_prize_9",
           "consolation_prize_10",
           "day_name"
           ]

df_data = pd.read_excel(DATASET_FILE, sheet_name="Damacai", header=None, names=COLUMNS)
display(df_data.head(15))

# Drop NaN
df_data.dropna(inplace = True)
display(df_data.head(15))
```

```python
# Transpose the dataset
def pivot_data(df):
    df_tabular = df.melt(
        id_vars=["year", "month", "day", "day_name"], var_name="prize_type", value_name="lucky_no")
    df_tabular = df_tabular.sort_values(["year", "month", "day", "day_name"], ascending=True)
    df_tabular.reset_index(inplace=True, drop=True)
    df_tabular["draw_date"] = df_tabular["year"].apply(str) + "-" + df_tabular["month"].apply(str) + "-" + df_tabular["day"].apply(str)
    df_tabular["draw_date"] = pd.to_datetime(df_tabular["draw_date"], format="%Y-%m-%d")
    return df_tabular

df_transposed = pivot_data(df_data)
df_transposed.head(10)
```

```python
# Convert data type
df_transposed["lucky_no"] = df_transposed["lucky_no"].apply(int)
df_transposed["day_name"] = df_transposed["day_name"].astype("category")
#df_transposed["lucky_no"] = df_transposed["lucky_no"].astype("category")
df_transposed.replace({"first_prize": 1,
                       "second_prize":1,
                       "third_prize": 1,
                       "starter_prize_1": 0,
                       "starter_prize_2": 0,
                       "starter_prize_3": 0,
                       "starter_prize_4": 0,
                       "starter_prize_5": 0,
                       "starter_prize_6": 0,
                       "starter_prize_7": 0,
                       "starter_prize_8": 0,
                       "starter_prize_9": 0,
                       "starter_prize_10": 0,
                       "consolation_prize_1": 0,
                       "consolation_prize_2": 0,
                       "consolation_prize_3": 0,
                       "consolation_prize_4": 0,
                       "consolation_prize_5": 0,
                       "consolation_prize_6": 0,
                       "consolation_prize_7": 0,
                       "consolation_prize_8": 0,
                       "consolation_prize_9": 0,
                       "consolation_prize_10": 0,
                      },
                       inplace=True)
df_transposed["prize_type"] = df_transposed["prize_type"].apply(int)
df_transposed.dtypes
```

# Training

```python
X = df_transposed[["draw_date", "lucky_no"]].copy()
y = df_transposed[["prize_type"]].copy()
print(len(X), len(y))
print(X.head())
print(y.head())
```

```python
def extract_digit(no, pos):
  no = str(no).zfill(4)
  return int(no[pos-1])

X["first"] = X["lucky_no"].apply(extract_digit, args=[1])
X["second"] = X["lucky_no"].apply(extract_digit, args=[2])
X["third"] = X["lucky_no"].apply(extract_digit, args=[3])
X["fourth"] = X["lucky_no"].apply(extract_digit, args=[4])

print(X.head())
print(y.head())
```
